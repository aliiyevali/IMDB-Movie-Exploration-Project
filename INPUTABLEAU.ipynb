```python
import numpy as np
import pandas as pd
inputcsv=pd.read_csv("/Users/aa085/Movies/Plo.csv")
inputcsv
```

```python
NList=inputcsv.columns
inputcsv.groupby("Action").sum()["wordlwide"][1]
```

```python
NList=list(NList)
NList=NList[22:]
NList
```

```python
inputcsv.groupby("War").sum()["IMDB"][1]
```

```python
TMList=[]
for a in NList:
    TMList.append(inputcsv.groupby(a).sum()["TMDB"][1])
TMList
```

```python
IMList=[]
for a in NList:
     IMList.append(inputcsv.groupby(a)['IMDB'].mean().loc[1])
IMList
```

```python
WList=[]
for a in NList:
    WList.append(inputcsv.groupby(a).sum()["wordlwide"][1])
WList
```

```python
Counts=[]
for a in NList:
    Counts.append(inputcsv[a].value_counts()[1])
Counts
```

```python
NList=pd.DataFrame(NList)
WList=pd.DataFrame(WList)
IMList=pd.DataFrame(IMList)
TMList=pd.DataFrame(TMList)
Counts=pd.DataFrame(Counts)
Compare=pd.concat([IMList,TMList,NList,WList,Counts],axis=1)
Compare
```

```python
Compare.columns = ["AVGIMDB","TotalTmdb",'genre', 'profit',"Counts"]
Compare
```

```python
inputcsv=inputcsv.drop(["BestActorPresent","Fouractors","NOactor","OneActor","Threeactors","Twoactors"],axis=1)
```

```python
Genre=["Adventure","Action","Comedy","Drama","Fantasy","Sci-Fi","Thriller","Family"]
for i,a in enumerate(Genre):
    Boxinput2=inputcsv[inputcsv[a]==1]
    Boxinput2=Boxinput2[["wordlwide",a]]
    Boxinput1=pd.concat([Boxinput1,Boxinput2],axis=0)
Boxinput1
```
